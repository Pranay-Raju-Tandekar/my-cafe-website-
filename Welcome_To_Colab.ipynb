Write a program that checks a temperature variable and prints "It's too hot!" if it's above 30, "The weather is perfect." if it's between 15 and 30, and "It's freezing!" otherwise.

```
# This is formatted as code
```



````
temperature = 33

if temperature > 30:
    print("It's too hot!")
elif 15 <= temperature <= 30:
    print("The weather is perfect.")
else:
    print("It's freezing!")
````

````
marks = 85

if marks >= 90:
    print("Grade: A+")
elif marks >= 75:
    print("Grade: A")  # This will run because 85 is greater than 75
elif marks >= 50:
    print("Grade: B")
else:
    print("Fail")
````
